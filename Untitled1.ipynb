### Introduction to Generative Adversarial Networks (GANs)

Generative Adversarial Networks (GANs) are a class of deep learning models designed to generate new data instances that resemble the training data. A GAN consists of two main components:

1.  **Generator (G)**: This network learns to create data. It takes a random noise vector as input and outputs data (e.g., an image) that tries to mimic the real data distribution.
2.  **Discriminator (D)**: This network learns to distinguish between real data (from the training set) and fake data (generated by the Generator). It acts as a binary classifier, outputting the probability that a given input is real.

The two networks are trained simultaneously in a zero-sum game: the Generator tries to fool the Discriminator by producing increasingly realistic data, while the Discriminator tries to get better at distinguishing real from fake. This adversarial process drives both networks to improve until the Generator produces data that is indistinguishable from real data by the Discriminator.

In your case, you're looking to improve a GAN using attention strategies. We'll first set up a basic GAN architecture, and then we can explore integrating self-attention or cross-attention into it.

### Loss Functions and Optimizers

GAN training involves a minimax game where the generator tries to minimize the discriminator's ability to distinguish real from fake images, and the discriminator tries to maximize its accuracy in distinguishing them. We'll use binary cross-entropy for the loss functions and Adam optimizers for both networks.

```python
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss

def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)

generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)
```

### Self-Attention Layer

Self-attention allows the model to weigh the importance of different parts of the input sequence (or feature map, in the case of images) when producing an output. For a GAN, this can help the generator to create more globally coherent and detailed images by attending to relevant features across the entire image. We will implement a basic self-attention mechanism to be integrated into our generator.

```python
class SelfAttention(layers.Layer):
    def __init__(self, ch, **kwargs):
        super(SelfAttention, self).__init__(**kwargs)
        self.ch = ch

        self.f = layers.Conv2D(ch // 8, 1, strides=1, padding='same')
        self.g = layers.Conv2D(ch // 8, 1, strides=1, padding='same')
        self.h = layers.Conv2D(ch, 1, strides=1, padding='same')
        self.o = layers.Conv2D(ch, 1, strides=1, padding='same')

        self.gamma = tf.Variable(tf.zeros(1))

    def call(self, x):
        shape = tf.shape(x)
        batch_size, h, w, ch = shape[0], shape[1], shape[2], shape[3]

        f_x = tf.reshape(self.f(x), [batch_size, h * w, ch // 8])
        g_x = tf.reshape(self.g(x), [batch_size, h * w, ch // 8])
        h_x = tf.reshape(self.h(x), [batch_size, h * w, ch])

        s = tf.matmul(g_x, f_x, transpose_b=True) # [batch_size, HW, HW]
        beta = tf.nn.softmax(s, axis=-1)

        o = tf.matmul(beta, h_x) # [batch_size, HW, C]
        o = tf.reshape(o, [batch_size, h, w, ch])

        return self.gamma * self.o(o) + x
```

### Training Loop

The training loop is where the magic happens. We'll iterate over epochs, and within each epoch, we'll train the discriminator and generator. The discriminator is trained first to classify real and fake images, then the generator is trained to produce more convincing fake images to fool the discriminator.

```python
EPOCHS = 50
BATCH_SIZE = 128
noise_dim = LATENT_DIM
num_examples_to_generate = 16

# We'll reuse this seed overtime (so it's easier to visualize progress)
seed = tf.random.normal([num_examples_to_generate, noise_dim])

# Load and preprocess the MNIST dataset
(train_images, train_labels), (_, _) = tf.keras.datasets.mnist.load_data()

train_images = train_images.reshape(train_images.shape[0], IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS).astype('float32')
# Normalize the images to be in the range [-1, 1]
train_images = (train_images - 127.5) / 127.5

train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(60000).batch(BATCH_SIZE)

@tf.function
def train_step(images):
    noise = tf.random.normal([BATCH_SIZE, noise_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator(noise, training=True)

        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

def generate_and_save_images(model, epoch, test_input):
  # Notice `training` is set to false. This is so all layers run in inference mode (batchnorm).
  predictions = model(test_input, training=False)

  fig = plt.figure(figsize=(4, 4))

  for i in range(predictions.shape[0]):
      plt.subplot(4, 4, i+1)
      plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
      plt.axis('off')

  # Create a directory for saving images if it doesn't exist
  if not os.path.exists('generated_images'):
      os.makedirs('generated_images')

  plt.savefig('generated_images/image_at_epoch_{:04d}.png'.format(epoch))
  plt.show()

def train(dataset, epochs):
    for epoch in range(epochs):
        for image_batch in dataset:
            train_step(image_batch)

        display.clear_output(wait=True)
        generate_and_save_images(generator, epoch + 1, seed)

        print (f'Epoch {epoch + 1} finished')

    # Generate after the final epoch
    display.clear_output(wait=True)
    generate_and_save_images(generator, epochs, seed)

print("Starting GAN training...")
# Train the GAN
train(train_dataset, EPOCHS)
print("GAN training complete.")
```

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np
import matplotlib.pyplot as plt
import os # Added for saving images
from IPython import display # Added for display.clear_output

# Define hyperparameters for the GAN (these can be tuned later)
LATENT_DIM = 100 # Dimension of the noise vector
IMG_HEIGHT = 28 # Example for MNIST
IMG_WIDTH = 28
IMG_CHANNELS = 1
```

### Generator Model

The Generator takes a latent space vector (random noise) as input and transforms it through several layers (e.g., dense layers, reshape, convolutional layers) to produce an image. The goal is to generate images that resemble the real training data.

```python
def make_generator_model():
    model = tf.keras.Sequential()
    model.add(tf.keras.Input(shape=(LATENT_DIM,)))
    model.add(layers.Dense(7*7*256, use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((7, 7, 256)))
    assert model.output_shape == (None, 7, 7, 256) # Note: None is the batch size

    model.add(layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False))
    assert model.output_shape == (None, 7, 7, 128)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    assert model.output_shape == (None, 14, 14, 64)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(IMG_CHANNELS, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert model.output_shape == (None, IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS)

    return model

# Test the generator
generator = make_generator_model()
noise = tf.random.normal([1, LATENT_DIM])
generated_image = generator(noise, training=False)

print("Generator output shape:", generated_image.shape)
# plt.imshow(generated_image[0, :, :, 0], cmap='gray')
# plt.title("Initial Generated Image (Noise)")
# plt.axis('off')
# plt.show()
```

### Discriminator Model

The Discriminator takes an image (either real from the dataset or fake from the Generator) as input and outputs a single scalar prediction indicating the probability that the image is 'real'. It's essentially a binary classifier.

```python
def make_discriminator_model():
    model = tf.keras.Sequential()
    model.add(tf.keras.Input(shape=[IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS]))
    model.add(layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    # Add Self-Attention Layer here, for example, after the first convolutional block
    model.add(SelfAttention(64)) # Apply attention to the 64-channel feature map

    model.add(layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Flatten())
    model.add(layers.Dense(1))

    return model

# Test the discriminator
discriminator = make_discriminator_model()
decision = discriminator(generated_image)

print("Discriminator output (raw logits):"), decision.numpy()
```
